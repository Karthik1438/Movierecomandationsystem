```python
import numpy as np
import pandas as pd
#stemmin class from nltk
from nltk.stem.porter import PorterStemmer
#count vectorizor
from sklearn.feature_extraction.text import CountVectorizer
#cosine similarity to calculate that similarity measure between movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Reading the dataset
df=pd.read_csv(r"C:\Users\karth\Downloads\movies_dataset.csv")
```

```python
ps=PorterStemmer()
```

```python
# Defining stemmin class
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)
```

```python
ps.stem('programmers')
```

```python
#Example of how words are reduced through stemming
# choose some words to be stemmed
#stemming example. A stemmeing algorithm reduces the words to the root/base word.
words = ["program", "programs", "programmer", "programming", "programmers"]
 
for w in words:
    print(w, " : ", ps.stem(w))
```

```python
'''for larger datasets ".apply(stem)" is more faster than Iterate over rows.
'''
#df['tags']=df['tags'].apply(stem)
```

```python
# Iterate over rows.
for index, row in df.iterrows():
    print(index)  # (optional debug line)
    df.loc[index, 'tags'] = stem(row['tags'])
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
# It prints all the reduced words
cv.get_feature_names_out()
```

```python
# Calculating cosine similarity
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_list_df=df[df['title'].str.contains(movie)]
    if len(movie_list_df):
        movie_index=movie_list_df.index[0]
        distance=similarity[movie_index]
        movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[0:6]
        print('The recommendations for {0} :\n'.format(movie_list_df.iloc[0]['title']))
        for i in movie_list:
            print(df.iloc[i[0]].title)
    else:
        return "No Movies found"
```
